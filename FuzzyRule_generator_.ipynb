```
#Step 1 fuzzify dataset


X= np.arange(1100,1601,1)
Y=np.arange(100,601,1) 
Z=np.arange(2100,2601,1)


XB2=np.max(X)
XS2=np.min(X)
XCE=np.median(X)
XB1=((np.abs((XCE)-(XB2)))/2)+XCE
XS1=((np.abs((XS2)-(XCE)))/2)+XS2
#############
YB2=np.max(Y)
YS2=np.min(Y)
YCE=np.median(Y)
YB1=((np.abs((YCE)-(YB2)))/2)+YCE
YS1=((np.abs((YS2)-(YCE)))/2)+YS2
#########
ZB2=np.max(Z)
ZS2=np.min(Z)
ZCE=np.median(Z)
ZB1=((np.abs((ZCE)-(ZB2)))/2)+ZCE
ZS1=((np.abs((ZS2)-(ZCE)))/2)+ZS2
######
l=[XB2,YB2,ZB2]

l=max(l)
L=l+l
########

YS3=(YS2)-L
YS4=(YS3)-L
YB3=(YB2)+L
YB4=(YB3)+L
XS3=(XS2)-L
XS4=(XS3)-L
XB3=(XB2)+L
XB4=(XB3)+L

ZS3=(ZS2)-L
ZS4=(ZS3)-L
ZB3=(ZB2)+L
ZB4=(ZB3)+L
# Generate fuzzy membership functions
fuzxS2 = fuzz.trapmf(X, [XS4, XS3, XS2, XS1])
fuzxS1 = fuzz.trimf(X, [XS2, XS1,XCE])
fuzxM = fuzz.trimf(X, [XS1, XCE, XB1])
fuzxH1 = fuzz.trimf(X, [XCE,XB1,XB2])
fuzxH2 = fuzz.trapmf(X, [XB1, XB2,XB3, XB4])


fuzyS2 = fuzz.trapmf(Y, [YS4, YS3, YS2, YS1])
fuzyS1 = fuzz.trimf(Y, [YS2, YS1,YCE])
fuzyM = fuzz.trimf(Y, [YS1, YCE, YB1])
fuzyH1 = fuzz.trimf(Y, [YCE,YB1,YB2])
fuzyH2= fuzz.trapmf(Y, [YB1,YB2,YB3, YB4])


fuzZS2 = fuzz.trapmf(Z, [ZS4, ZS3, ZS2, ZS1])
fuzZS1 = fuzz.trimf(Z, [ZS2, ZS1,ZCE])
fuzZM = fuzz.trimf(Z, [ZS1, ZCE, ZB1])
fuzZH1 = fuzz.trimf(Z, [ZCE,ZB1,ZB2])
fuzZH2= fuzz.trapmf(Z, [ZB1,ZB2,ZB3, ZB4])
# Visualize these universes and membership functions
fig, (ax0, ax1, ax2) = plt.subplots(nrows=3, figsize=(8, 10))

ax0.plot(X, fuzxS2, 'b', linewidth=1.5, label='1')
ax0.plot(X, fuzxS1, 'b', linewidth=1.5, label='2')
ax0.plot(X, fuzxM, 'g', linewidth=1.5, label='3')
ax0.plot(X, fuzxH1, 'r', linewidth=1.5, label='4')
ax0.plot(X, fuzxH2, 'r', linewidth=1.5, label='5')
ax0.set_title('X')
ax0.legend()

ax1.plot(Y, fuzyS2, 'b', linewidth=1.5, label='1')
ax1.plot(Y, fuzyS1, 'b', linewidth=1.5, label='2')
ax1.plot(Y, fuzyM, 'g', linewidth=1.5, label='3')
ax1.plot(Y, fuzyH1, 'r', linewidth=1.5, label='4')
ax1.plot(Y, fuzyH2, 'r', linewidth=1.5, label='5')
ax1.set_title('Y')
ax1.legend()

ax2.plot(Z, fuzZS2, 'b', linewidth=1.5, label='1')
ax2.plot(Z, fuzZS1, 'b', linewidth=1.5, label='2')
ax2.plot(Z, fuzZM, 'g', linewidth=1.5, label='3')
ax2.plot(Z, fuzZH1, 'r', linewidth=1.5, label='4')
ax2.plot(Z, fuzZH2, 'r', linewidth=1.5, label='5')
ax2.set_title('Z')
ax2.legend()

# Turn off top/right axes
for ax in (ax0, ax1, ax2):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```

```
#step 2 GENERATINFGUZZY RULESFROM NUMERICADLATA
'''
x1=900 y1=320 z1=2500

'''
X_input=[1211,1366,1366,1211,1211,1366,1366,1211,1211,1366,1366,1211,1211,1366,1366,1211,1211,1366,1366,1211,1211,1366,1366,1211,1400,1100,1290]
Y_input=[260,532,532,260,260,532,532,260,260,532,532,260,260,532,532,260,260,532,532,260,260,532,532,260,350,300,489]
Z_input=[2310,2467,2467,2245,2310,2467,2467,2245,2310,2467,2467,2245,2310,2467,2467,2245,2310,2467,2467,2245,2310,2467,2467,2245,2100,2300,2460]
Rules=[]
for i in range(len(X_input)):
  data_array=[X_input[i],Y_input[i],Z_input[i]]
  Xnm=[data_array[0]]
  Ynm=[data_array[1]]
  Znm=[data_array[2]]
  Xnma=np.array(Xnm)
  Ynma=np.array(Ynm)
  Znma=np.array(Znm)
  fuzxnmS2 = fuzz.trapmf(Xnma, [XS4, XS3, XS2, XS1])
  fuzxnmS1 = fuzz.trimf(Xnma, [XS2, XS1,XCE])
  fuzxnmM = fuzz.trimf(Xnma, [XS1, XCE, XB1])
  fuzxnmH1 = fuzz.trimf(Xnma, [XCE,XB1,XB2])
  fuzxnmH2 = fuzz.trapmf(Xnma, [XB1, XB2,XB3, XB4])
  mu_X=[fuzxnmS2,fuzxnmS1,fuzxnmM,fuzxnmH1,fuzxnmH2]
  fuzynmS2 = fuzz.trapmf(Ynma, [YS4, YS3, YS2, YS1])
  fuzynmS1 = fuzz.trimf(Ynma, [YS2, YS1,YCE])
  fuzynmM = fuzz.trimf(Ynma, [YS1, YCE, YB1])
  fuzynmH1 = fuzz.trimf(Ynma, [YCE,YB1,YB2])
  fuzynmH2= fuzz.trapmf(Ynma, [YB1,YB2,YB3, YB4])
  mu_Y=[fuzynmS2,fuzynmS1,fuzynmM,fuzynmH1,fuzynmH2]
  fuzZnmS2 = fuzz.trapmf(Znma, [ZS4, ZS3, ZS2, ZS1])
  fuzZnmS1 = fuzz.trimf(Znma, [ZS2, ZS1,ZCE])
  fuzZnmM = fuzz.trimf(Znma, [ZS1, ZCE, ZB1])
  fuzZnmH1 = fuzz.trimf(Znma, [ZCE,ZB1,ZB2])
  fuzZnmH2= fuzz.trapmf(Znma, [ZB1,ZB2,ZB3, ZB4])
  mu_Z=[fuzZnmS2,fuzZnmS1,fuzZnmM,fuzZnmH1,fuzZnmH2]
  #print(mu_X)
  index_mu_X= np.nonzero(mu_X)
  index_max_valu_mu_X=np.argmax(mu_X)
  max_valu_mu_X=mu_X[index_max_valu_mu_X]
  # for i in range(index_mu_X[0][0],index_mu_X[0][1]+1):
  # A=mu_X[i]
  # print('x in phase',(i),'and membrship value',A)
  #print(mu_Y)
  index_mu_Y= np.nonzero(mu_Y)
  index_max_valu_mu_Y=np.argmax(mu_Y)
  max_valu_mu_Y=mu_Y[index_max_valu_mu_Y]
  # for i in range(index_mu_Y[0][0],index_mu_Y[0][1]+1):
  # B=mu_Y[i]
  # print('y in phase',(i),'and membrship value',B)
  #print(mu_Z)
  index_mu_Z= np.nonzero(mu_Z)
  index_max_valu_mu_Z=np.argmax(mu_Z)
  max_valu_mu_Z=mu_Z[index_max_valu_mu_Z]
  # for i in range(index_mu_Z[0][0],index_mu_Z[0][1]+1):
  # A=mu_Z[i]
  # print('Z in phase',(i),'and membrship value',A)
  degree_rule=float((max_valu_mu_X)*(max_valu_mu_Y)*(max_valu_mu_Z))
  Rule=[index_max_valu_mu_X+1,index_max_valu_mu_Y+1,index_max_valu_mu_Z+1,degree_rule]
  print('The Rule is IF X is in phase',Rule[0],'and Y is in phase',Rule[1],'Then Z(result) is in phase',Rule[2],'Degree of this Rule =',Rule[3])
  Rules.append(Rule)
print(Rules) 
```
